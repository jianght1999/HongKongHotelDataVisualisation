```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
data_path = r"D:\codelearning\dataanalysis\酒店数据.xlsx"
data = pd.read_excel(data_path)
```

```python
print(data.head(5))
```

```python
#香港各个地区的酒店数量
region_counts = data['地区'].value_counts()
colors = plt.cm.Blues(0.5*region_counts.values / float(max(region_counts.values)))
plt.bar(region_counts.index, region_counts.values, color = colors)
#更换字体为微软雅黑
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
#添加文本标签
for x, y in zip(region_counts.index, region_counts.values):
    plt.text(x, y, str(y), ha='center', va='bottom')
plt.title('各个地区酒店分布')
plt.xlabel('地区')
plt.ylabel('数量')
plt.xticks(rotation=45)

plt.show()
```
