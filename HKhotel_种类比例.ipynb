```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
data_path = r"D:\codelearning\dataanalysis\酒店数据.xlsx"
data = pd.read_excel(data_path)
```

```python
print(data.head(5))
```

```python
counts = data['类型'].value_counts()
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']

# 选择出现次数最多的四个标签，并将其他标签合并为一个“其他类型”的标签
top_labels = counts.nlargest(4)
other_count = counts[~counts.index.isin(top_labels.index)].sum()
top_labels['其他类型'] = other_count

plt.pie(top_labels, labels=top_labels.index, autopct='%1.1f%%')
plt.title('酒店类型占比')
plt.axis('equal')
plt.show()
```
